#Load database - lightning

```python
import sqlite3

my_connection=sqlite3.connect("lightning_data.db")

my_cursor=my_connection.cursor()
```

##get table names

```python
# Getting all tables from sqlite_master
query_result = my_cursor.execute("SELECT name FROM sqlite_master WHERE type='table'")

print("List of tables\n")

# printing all tables list
print(query_result.fetchall())
```

##X_table

```python
query_result = my_cursor.execute("SELECT * from X_table")

for row in query_result:
  print(row)
```

```python
data_X = my_cursor.execute("SELECT * from X_table").fetchall()
data_X
```

```python
my_column_names=[]

for column in query_result.description:
    print(column[0])
    my_column_names.append(column[0])

print('-----------')
my_column_names
```

```python
import pandas as pd

df_X = pd.DataFrame(data_X, columns = my_column_names)

df_X
```

##y_table

```python
data_y = my_cursor.execute("SELECT * from y_table").fetchall()
data_y
```

```python
my_column_names=[]

for column in query_result.description:
    print(column[0])
    my_column_names.append(column[0])

print('-----------')
my_column_names
```

```python
df_y = pd.DataFrame(data_y, columns = my_column_names)

df_y
```

##Merge into one dataframe while fixing sorting problem

```python
df_all = pd.merge(df_X, df_y, on='Record_ID')

#don't need RECORD_ID anymore
df_all.drop(['Record_ID'], axis=1, inplace=True)
df_all
```

##Split into X & y

```python
import statsmodels.api as sm

X = df_all.loc[:, df_all.columns != 'OCCURRED']
X = sm.add_constant(X)

y = df_all.loc[:, df_all.columns == 'OCCURRED']

X
```

```python
y
```

##Model using logistic regression

```python
model=sm.Logit(y,X).fit()
print(model.summary())
```

```python
from sklearn import metrics

#Confusion Matrix
y_pred = model.predict(X) > 0.5
confusion_matrix = metrics.confusion_matrix(y, y_pred)

print("\nConfusion Matrix")
print(confusion_matrix)


#Classification Report
print("\nClassification Report")
from sklearn.metrics import classification_report
print(classification_report(y, y_pred, digits=2))
```
